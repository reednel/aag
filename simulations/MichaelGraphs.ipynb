```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('many.csv')
df['efficiency'] = df['attack'] / df['exchange']
```

```python
sns.set_theme(style="whitegrid")
gfg = sns.pairplot(data=df, hue='pk', palette='crest')
plt.show()
```

```python
sns.set_theme(style="whitegrid")
gfg = sns.pairplot(data=df, hue='sk', palette='flare')
plt.show()
```

```python
sns.set_theme(style="whitegrid")
gfg = sns.pairplot(data=df[df['pk']==7], hue='sk', vars=['exchange', 'attack'], palette='flare')
plt.show()
```

```python
sns.set_theme(style="whitegrid")
gfg = sns.pairplot(data=df, hue='pk', vars=['exchange', 'attack'], palette='crest')
plt.show()
```

```python
sns.set_theme(style="whitegrid")
gfg = sns.jointplot(data=df, x='exchange', y='attack', hue='sk', palette='flare')
plt.show()
```

```python
sns.set_theme(style="whitegrid")
gfg = sns.scatterplot(data=df, x='exchange', y='attack', hue='sk', palette='flare')
plt.show()
```

```python
sns.set_theme(style="whitegrid")
gfg = sns.kdeplot(
   data=df[(df.group=='S16') & (df.pk==5) & (df.sk <= 8)], x="attack", hue="sk",
   fill=True, common_norm=False, palette="flare",
   alpha=.5, linewidth=0, log_scale=True
)
gfg.set(title="Attack Time vs Private Key Length (S16, |pk|=5)", xlabel="Attack Time (ms)")
plt.show()
```

```python
sns.set_theme(style="whitegrid")
gfg = sns.kdeplot(
   data=df[(df.group=='S16') & (df.pk==5) & (df.sk <= 8)], x="guesses", hue="sk",
   fill=True, common_norm=False, palette="flare",
   alpha=.5, linewidth=0, log_scale=True
)
gfg.set(title="Brute-force Guesses vs Private Key Length (S16, |pk|=5)", xlabel="Guesses")
plt.show()
```

```python
sns.set_theme(style="whitegrid")
df2 = df
gfg = sns.kdeplot(
   data=df[(df.group=='S16') & (df.sk==5)], x="guesses", hue="pk",
   fill=True, common_norm=False, palette="crest",
   alpha=.5, linewidth=0, log_scale=True
)
gfg.set(title="Attack Time vs Public Key Length", xlabel="Attack Time (ms)")
plt.show()
```

```python
# heatmap comparing sk and pk effects on attack time
sns.set_theme(style="whitegrid")
df['efficiency'] = df['attack'] / df['exchange']
g = sns.JointGrid(data=df, x="pk", y="sk")
g.plot_joint(sns.scatterplot, c=df["efficiency"], size=df["attack"], sizes=(30,1200), palette="flare", alpha=.6, legend=False)
g.plot_marginals(sns.histplot, color="m", kde=True)
plt.show()
```

```python
# on the same axes, plot the attack time vs exchange time
sns.set_theme(style="whitegrid")
g = sns.JointGrid(data=df, x="efficiency", y="sk")
g.plot_joint(sns.scatterplot, c=df["efficiency"], size=df["attack"], sizes=(30,1200), palette="flare", alpha=.6, legend=False)
g.plot_marginals(sns.histplot, color="m", kde=True)
plt.show()
```
