```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('permutation_all.csv')
df['attack'] *= 1e-3
df['exchange'] *= 1e-3
```

```python
sns.set_theme(style="whitegrid")
gfg = sns.pairplot(data=df, hue='pk', palette='crest')
plt.show()
```

```python
sns.set_theme(style="whitegrid")
gfg = sns.pairplot(data=df, hue='sk', palette='flare')
plt.show()
```

```python
sns.set_theme(style="whitegrid")
gfg = sns.pairplot(data=df, hue='sk', vars=['exchange', 'attack'], palette='flare')
plt.show()
```

```python
sns.set_theme(style="whitegrid")
gfg = sns.pairplot(data=df, hue='pk', vars=['exchange', 'attack'], palette='crest')
plt.show()
```

```python
sns.set_theme(style="whitegrid")
gfg = sns.jointplot(data=df, x='exchange', y='attack', hue='sk', palette='flare')
plt.show()
```

```python
sns.set_theme(style="whitegrid")
gfg = sns.scatterplot(data=df, x='exchange', y='attack', hue='sk', palette='flare')
plt.show()
```

```python
sns.set_theme(style="whitegrid")
gfg = sns.kdeplot(
   data=df, x="attack", hue="sk",
   fill=True, common_norm=False, palette="flare",
   alpha=.5, linewidth=0, log_scale=True
)
gfg.set(title="Attack Time vs Private Key Length", xlabel="Attack Time (ms)")
plt.show()
```

```python
sns.set_theme(style="whitegrid")
df2 = df[df['sk'] == 6]
gfg = sns.kdeplot(
   data=df2, x="attack", hue="pk",
   fill=True, common_norm=False, palette="crest",
   alpha=.5, linewidth=0, log_scale=True
)
gfg.set(title="Attack Time vs Public Key Length", xlabel="Attack Time (ms)")
plt.show()
```
